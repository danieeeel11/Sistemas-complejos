```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
plt.style.use('dark_background')
```

```python
# masses of planets
m_1 = 10
m_2 = 20

# starting coordinates for planets
p1_start = np.array([-10, 10, -11])
v1_start = np.array([-3, 0, 0])

p2_start = np.array([0, 0, 0])
v2_start = np.array([0, 0, 0])
```

```python
def accelerations(p1, p2):
    # Calculate derivatives of x, y, and z according to Newton's laws
    planet_1_dv = -9.8 * m_2 * (p1 - p2) / (np.sqrt(np.sum((p1 - p2)**2))**3)
    planet_2_dv = -9.8 * m_1 * (p2 - p1) / (np.sqrt(np.sum((p2 - p1)**2))**3)
    return planet_1_dv, planet_2_dv
```

```python
# parameters
delta_t = 0.001
steps = 200000

# initialize trajectory array
p1 = np.zeros((steps, 3))
v1 = np.zeros((steps, 3))

p2 = np.zeros((steps, 3))
v2 = np.zeros((steps, 3))
```

```python
# starting point and velocity
p1[0], p2[0] = p1_start, p2_start
v1[0], v2[0] = v1_start, v2_start
```

```python
# evolution of the system
for i in range(steps-1):
    dv1, dv2 = accelerations(p1[i], p2[i])

    v1[i + 1] = v1[i] + dv1 * delta_t
    v2[i + 1] = v2[i] + dv2 * delta_t

    p1[i + 1] = p1[i] + v1[i + 1] * delta_t
    p2[i + 1] = p2[i] + v2[i + 1] * delta_t
```

```python
# Gráfica mejorada
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
plt.gca().patch.set_facecolor('black')

ax.plot(p1[:, 0], p1[:, 1], p1[:, 2], '^', color='red', lw=0.05, markersize=0.01, alpha=0.5)
ax.plot(p2[:, 0], p2[:, 1], p2[:, 2], '^', color='white', lw=0.05, markersize=0.01, alpha=0.5)

# Configuración de la vista
ax.mouse_init()
ax.view_init(elev=20, azim=30)  # Puedes ajustar estos valores según tus preferencias

# Agregar leyendas
ax.legend(['Masa 1', 'Masa 2'])

plt.show()
plt.close()
```

### Explicacion

Cuando hay solo dos cuerpos masivos interactuando entre sí bajo la influencia de la gravedad, el problema se simplifica considerablemente en comparación con el problema de los tres cuerpos. Este escenario se conoce como el "problema de los dos cuerpos" y es más fácil de resolver analíticamente. Aquí están las principales características y resultados del problema de los dos cuerpos:


* Formulación del problema: En el problema de los dos cuerpos, solo hay dos objetos masivos que interactúan entre sí gravitacionalmente. Por ejemplo, esto podría ser dos estrellas, dos planetas, o un planeta y una luna.


* Ecuaciones de movimiento: Las ecuaciones de movimiento en el problema de los dos cuerpos se derivan de la segunda ley de Newton y la ley de gravitación universal de Newton. Estas ecuaciones resultan en ecuaciones diferenciales que describen la trayectoria de ambos cuerpos en función de sus posiciones iniciales, sus masas y las fuerzas gravitatorias que actúan entre ellos.


* Solución analítica: A diferencia del problema de los tres cuerpos, el problema de los dos cuerpos tiene una solución analítica exacta. Para el caso de dos cuerpos interactuantes bajo la influencia exclusiva de la gravedad, la trayectoria de ambos cuerpos sigue una forma cónica, como una elipse, una parábola o una hipérbola, dependiendo de las condiciones iniciales y las energías relativas.


* Leyes de Kepler: Las soluciones del problema de los dos cuerpos siguen las leyes de Kepler del movimiento planetario, que son tres leyes empíricas formuladas por Johannes Kepler en el siglo XVII. Estas leyes describen la forma de las órbitas elípticas de los planetas alrededor del Sol, así como la relación entre el período orbital y la distancia al centro de masas.


En resumen, cuando hay solo dos cuerpos masivos interactuando gravitacionalmente entre sí, el problema se simplifica y se puede resolver analíticamente. Esto conduce a una descripción precisa del movimiento de los cuerpos a través de soluciones cónicas, como elipses, parábolas o hipérbolas, de acuerdo con las leyes de Kepler del movimiento planetario.

```python
# masses of planets
m_1 = 10
m_2 = 20
m_3 = 30

# starting coordinates for planets
# p1_start = x_1, y_1, z_1
p1_start = np.array([-10, 10, -11])
v1_start = np.array([-3, 0, 0])

# p2_start = x_2, y_2, z_2
p2_start = np.array([0, 0, 0])
v2_start = np.array([0, 0, 0])

# p3_start = x_3, y_3, z_3
p3_start = np.array([10, 10, 12])
v3_start = np.array([3, 0, 0])
```

```python
def accelerations(p1, p2, p3):
	"""
	A function to calculate the derivatives of x, y, and z
	given 3 object and their locations according to Newton's laws

	"""

	#m_1, m_2, m_3 = m1, m2, m3
	planet_1_dv = -9.8 * m_2 * (p1 - p2)/(np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)**3) - \
		       9.8 * m_3 * (p1 - p3)/(np.sqrt((p1[0] - p3[0])**2 + (p1[1] - p3[1])**2 + (p1[2] - p3[2])**2)**3)

	planet_2_dv = -9.8 * m_3 * (p2 - p3)/(np.sqrt((p2[0] - p3[0])**2 + (p2[1] - p3[1])**2 + (p2[2] - p3[2])**2)**3) - \
		       9.8 * m_1 * (p2 - p1)/(np.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2 + (p2[2] - p1[2])**2)**3)

	planet_3_dv = -9.8 * m_1 * (p3 - p1)/(np.sqrt((p3[0] - p1[0])**2 + (p3[1] - p1[1])**2 + (p3[2] - p1[2])**2)**3) - \
		       9.8 * m_2 * (p3 - p2)/(np.sqrt((p3[0] - p2[0])**2 + (p3[1] - p2[1])**2 + (p3[2] - p2[2])**2)**3)

	return planet_1_dv, planet_2_dv, planet_3_dv
```

```python
# parameters
delta_t = 0.001
steps = 200000

# initialize trajectory array
p1 = np.array([[0.,0.,0.] for i in range(steps)])
v1 = np.array([[0.,0.,0.] for i in range(steps)])

p2 = np.array([[0.,0.,0.] for j in range(steps)])
v2 = np.array([[0.,0.,0.] for j in range(steps)])

p3 = np.array([[0.,0.,0.] for k in range(steps)])
v3 = np.array([[0.,0.,0.] for k in range(steps)])
```

```python
# starting point and velocity
p1[0], p2[0], p3[0] = p1_start, p2_start, p3_start

v1[0], v2[0], v3[0] = v1_start, v2_start, v3_start
```

```python
# evolution of the system
for i in range(steps-1):
	# calculate derivatives
	dv1, dv2, dv3 = accelerations(p1[i], p2[i], p3[i])

	v1[i + 1] = v1[i] + dv1 * delta_t
	v2[i + 1] = v2[i] + dv2 * delta_t
	v3[i + 1] = v3[i] + dv3 * delta_t

	p1[i + 1] = p1[i] + v1[i] * delta_t
	p2[i + 1] = p2[i] + v2[i] * delta_t
	p3[i + 1] = p3[i] + v3[i] * delta_t
```

```python
# Gráfica mejorada
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
plt.gca().patch.set_facecolor('black')

ax.plot(p1[:, 0], p1[:, 1], p1[:, 2], '^', color='red', lw=0.05, markersize=0.01, alpha=0.5)
ax.plot(p2[:, 0], p2[:, 1], p2[:, 2], '^', color='white', lw=0.05, markersize=0.01, alpha=0.5)
ax.plot(p3[:, 0], p3[:, 1], p3[:, 2], '^', color='blue', lw=0.05, markersize=0.01, alpha=0.5)

# Configuración de la vista
ax.mouse_init()
ax.view_init(elev=20, azim=30)  # Puedes ajustar estos valores según tus preferencias

# Agregar leyendas
ax.legend(['Masa 1', 'Masa 2', 'Masa 3', 'Masa 4'])

plt.show()
plt.close()
```

### Explicacion

Cuando hay tres cuerpos masivos interactuando gravitacionalmente entre sí, como en el problema de los tres cuerpos, la situación se vuelve considerablemente más compleja en comparación con el problema de los dos cuerpos. En este caso, surge una serie de fenómenos que hacen que la predicción del comportamiento a largo plazo sea impredecible o muy difícil de calcular con precisión. Algunos de estos fenómenos incluyen:


* Caos y sensibilidad a las condiciones iniciales: El problema de los tres cuerpos exhibe un comportamiento caótico, lo que significa que pequeñas variaciones en las condiciones iniciales pueden dar lugar a resultados radicalmente diferentes a largo plazo. Esto se debe a la naturaleza no lineal de las ecuaciones de movimiento y a la interacción compleja entre los cuerpos. Como resultado, la predicción precisa del movimiento a largo plazo se vuelve extremadamente difícil, si no imposible.


* Soluciones no periódicas: A diferencia del problema de los dos cuerpos, donde las órbitas siguen trayectorias cónicas predecibles (elipses, parábolas o hipérbolas), en el problema de los tres cuerpos las trayectorias no siguen un patrón simple y pueden ser altamente irregulares. Esto puede llevar a que los cuerpos no sigan órbitas cerradas y periódicas, sino que exhiban movimientos más complejos y aparentemente caóticos.


* Puntos de equilibrio: En el problema de los tres cuerpos, existen puntos especiales llamados puntos de equilibrio o puntos lagrangianos, donde las fuerzas gravitatorias entre los cuerpos se equilibran. Estos puntos pueden tener estabilidad limitada y se utilizan en la mecánica celeste para colocar satélites artificiales en órbitas estables alrededor de cuerpos celestes.


* Simulaciones numéricas: Debido a la dificultad para obtener soluciones analíticas exactas, el estudio del problema de los tres cuerpos a menudo se realiza mediante simulaciones numéricas utilizando métodos computacionales avanzados. Estas simulaciones pueden proporcionar información detallada sobre el comportamiento a largo plazo de los sistemas de tres cuerpos y son ampliamente utilizadas en la astrofísica y la dinámica orbital.


En resumen, el problema de los tres cuerpos es altamente complejo y exhibe un comportamiento caótico que dificulta la predicción precisa del movimiento a largo plazo. Aunque se pueden identificar algunos patrones generales y puntos de equilibrio, la naturaleza no lineal de las ecuaciones de movimiento hace que sea imposible predecir el comportamiento exacto de los cuerpos en muchas situaciones.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

plt.style.use('dark_background')

# Masas de los planetas
m_1 = 15
m_2 = 25
m_3 = 20
m_4 = 30

# Coordenadas iniciales de los planetas
p1_start = np.array([-15, 10, -5])
v1_start = np.array([-1, 0.5, 0])

p2_start = np.array([5, -5, 8])
v2_start = np.array([0.5, -1, 0])

p3_start = np.array([10, 15, -12])
v3_start = np.array([1, 0, 0.8])

p4_start = np.array([-8, -10, 6])
v4_start = np.array([0, 0.8, -1])

def accelerations(p1, p2, p3, p4):
    dv1 = -9.8 * (m_2 * (p1 - p2) / np.sqrt(np.sum((p1 - p2)**2))**3 +
                  m_3 * (p1 - p3) / np.sqrt(np.sum((p1 - p3)**2))**3 +
                  m_4 * (p1 - p4) / np.sqrt(np.sum((p1 - p4)**2))**3)

    dv2 = -9.8 * (m_1 * (p2 - p1) / np.sqrt(np.sum((p2 - p1)**2))**3 +
                  m_3 * (p2 - p3) / np.sqrt(np.sum((p2 - p3)**2))**3 +
                  m_4 * (p2 - p4) / np.sqrt(np.sum((p2 - p4)**2))**3)

    dv3 = -9.8 * (m_1 * (p3 - p1) / np.sqrt(np.sum((p3 - p1)**2))**3 +
                  m_2 * (p3 - p2) / np.sqrt(np.sum((p3 - p2)**2))**3 +
                  m_4 * (p3 - p4) / np.sqrt(np.sum((p3 - p4)**2))**3)

    dv4 = -9.8 * (m_1 * (p4 - p1) / np.sqrt(np.sum((p4 - p1)**2))**3 +
                  m_2 * (p4 - p2) / np.sqrt(np.sum((p4 - p2)**2))**3 +
                  m_3 * (p4 - p3) / np.sqrt(np.sum((p4 - p3)**2))**3)

    return dv1, dv2, dv3, dv4

# Parámetros de la simulación
delta_t = 0.001
steps = 200000

# Inicialización de los arreglos de trayectorias
p1 = np.zeros((steps, 3))
v1 = np.zeros((steps, 3))

p2 = np.zeros((steps, 3))
v2 = np.zeros((steps, 3))

p3 = np.zeros((steps, 3))
v3 = np.zeros((steps, 3))

p4 = np.zeros((steps, 3))
v4 = np.zeros((steps, 3))

# Condiciones iniciales
p1[0], p2[0], p3[0], p4[0] = p1_start, p2_start, p3_start, p4_start
v1[0], v2[0], v3[0], v4[0] = v1_start, v2_start, v3_start, v4_start

# Evolución del sistema
for i in range(steps-1):
    dv1, dv2, dv3, dv4 = accelerations(p1[i], p2[i], p3[i], p4[i])

    v1[i + 1] = v1[i] + dv1 * delta_t
    v2[i + 1] = v2[i] + dv2 * delta_t
    v3[i + 1] = v3[i] + dv3 * delta_t
    v4[i + 1] = v4[i] + dv4 * delta_t

    p1[i + 1] = p1[i] + v1[i + 1] * delta_t
    p2[i + 1] = p2[i] + v2[i + 1] * delta_t
    p3[i + 1] = p3[i] + v3[i + 1] * delta_t
    p4[i + 1] = p4[i] + v4[i + 1] * delta_t
```
